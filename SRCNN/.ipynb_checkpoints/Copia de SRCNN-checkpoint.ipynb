```python
import os
import numpy as np
from PIL import Image

def create_average_image(directory):
    # Get a list of all subdirectories within the main directory
    subdirectories = [os.path.join(directory, d) for d in os.listdir(directory) if os.path.isdir(os.path.join(directory, d))]

    # Initialize variables for accumulating the pixel values
    total_images = 0
    average_image = None

    # Iterate through each subdirectory
    for subdirectory in subdirectories:
        # Get a list of TIFF image files within the subdirectory
        image_files = [f for f in os.listdir(subdirectory) if os.path.isfile(os.path.join(subdirectory, f)) and f.lower().endswith(".tif")]

        # Iterate through each TIFF image file
        for image_file in image_files:
          try:
            # Open the image using PIL
            image_path = os.path.join(subdirectory, image_file)
            image = Image.open(image_path)

            # Convert the image to a numpy array
            pixels = np.array(image)

            # Add the pixel values to the accumulated sum
            if average_image is None:
                average_image = pixels.astype(np.uint64)
            else:
                average_image += pixels.astype(np.uint64)

            # Update the total image count
            total_images += 1

          except Exception as e:
            continue

    # Calculate the average pixel values
    average_image = (average_image // total_images).astype(np.uint8)

    # Create a new PIL image from the average pixel values
    average_image = Image.fromarray(average_image)

    return average_image

# Example usage
directory_path = "/content/drive/MyDrive/Super Resolution/Sofia_experiments_8bit/Qc_0p10mLminQ_d0p07Lmin_freq1000Hz"
average_image = create_average_image(directory_path)
average_image.show()  # Display the average image
average_image.save("/content/drive/MyDrive/Super Resolution/Sofia_experiments_8bit/Qc_0p10mLminQ_d0p07Lmin_freq1000Hz.tif")  # Save the average image
```

```python
import os
from PIL import Image
import numpy as np

def convert_16bit_to_8bit(directory, output_directory):
    # Create the output directory if it doesn't exist
    os.makedirs(output_directory, exist_ok=True)

    # Get a list of all subdirectories within the main directory
    subdirectories = [os.path.join(directory, d) for d in os.listdir(directory) if os.path.isdir(os.path.join(directory, d))]

    # Iterate through each subdirectory
    for subdirectory in subdirectories:
        # Create the corresponding subdirectory in the output directory
        relative_path = os.path.relpath(subdirectory, directory)
        output_subdirectory = os.path.join(output_directory, relative_path)
        os.makedirs(output_subdirectory, exist_ok=True)

        # Get a list of TIFF image files within the subdirectory
        image_files = [f for f in os.listdir(subdirectory) if os.path.isfile(os.path.join(subdirectory, f)) and f.lower().endswith(".tif")]

        print("Saved " + subdirectory)

        # Iterate through each TIFF image file
        for image_file in image_files:
            # Open the image using PIL
            image_path = os.path.join(subdirectory, image_file)
            image = Image.open(image_path)

            # Convert the image to a numpy array
            pixels = np.array(image)

            # Linearly scale the pixel values from the minimum-maximum range to 0-255
            min_val = np.min(pixels)
            max_val = np.max(pixels)
            scaled_pixels = ((pixels - min_val) / (max_val - min_val) * 255).astype(np.uint8)

            # Create a new PIL image from the scaled pixel values
            image_8bit = Image.fromarray(scaled_pixels)

            # Save the 8-bit image in the corresponding output subdirectory
            output_path = os.path.join(output_subdirectory, image_file)
            image_8bit.save(output_path)


# Example usage
input_directory = "/content/drive/MyDrive/Super Resolution/Sofia_experiments"
output_directory = "/content/drive/MyDrive/Super Resolution/Sofia_experiments_8bit"
convert_16bit_to_8bit(input_directory, output_directory)
```

```python
import argparse
import glob
import numpy as np
import os
import PIL.Image as pil_image

def downsample_images(images_dir, output_dir, scale):
    os.makedirs(output_dir, exist_ok=True)

    image_files = sorted(glob.glob('{}/*.tif'.format(images_dir)))

    for image_path in image_files:
        image_name = os.path.basename(image_path)
        output_path = os.path.join(output_dir, image_name)

        hr = Image.open(image_path)
        width, height = hr.size

        hr_width = (width // scale) * scale
        hr_height = (height // scale) *scale

        hr = hr.resize((hr_width, hr_height), resample=pil_image.BICUBIC)
        lr = hr.resize((hr_width // scale, hr_height // scale), resample=pil_image.BICUBIC)
        lr = lr.resize((lr.width * scale, lr.height * scale), resample=pil_image.BICUBIC)

        lr.save(output_path)

        #print("Downsampled image saved:", output_path)
```

```python
import os
import numpy as np
from PIL import Image
import PIL.Image as pil_image

def create_average_image(directory, hr = False, scale = False):
    # Get a list of all image files within the directory
    image_files = [f for f in os.listdir(directory) if os.path.isfile(os.path.join(directory, f)) and f.lower().endswith(".tif")]

    # Initialize variables for accumulating the pixel values
    total_images = 0
    average_image = None

    # Iterate through each image file
    for image_file in image_files:
        # Open the image using PIL
        image_path = os.path.join(directory, image_file)
        hr = Image.open(image_path)
               
        hr_width = (hr.width // 4) * 4
        hr_height = (hr.height // 4) * 4
        hr = hr.resize((hr_width, hr_height), resample=pil_image.BICUBIC)

        image= hr

        # Convert the image to a numpy array
        pixels = np.array(image)

        # Add the pixel values to the accumulated sum
        if average_image is None:
            average_image = pixels.astype(np.uint64)
        else:
            average_image += pixels.astype(np.uint64)

        # Update the total image count
        total_images += 1

    # Calculate the average pixel values
    average_image = (average_image // total_images).astype(np.uint8)

    # Create a new PIL image from the average pixel values
    average_image = Image.fromarray(average_image)

    return average_image

# Example usage
directory_path = "/home/daniel/Documents/Super Resolution/Preprocessed/Cropped_Qc_0p10mLminQ_d0p05Lmin_freq1000Hz"
average_image = create_average_image(directory_path, True, 2)
average_image.show()  # Display the average image
average_image.save("/home/daniel/Documents/Super Resolution/Preprocessed/Cropped_Qc_0p10mLminQ_d0p05Lmin_freq1000Hz/Average_Downsampledx2_Qc_0p10mLminQ_d0p05Lmin_freq1000Hz.tif")  # Save the average image
```

```python
## Preprocessing

import os
from PIL import Image
import numpy as np
import PIL.Image as pil_image
import matplotlib.pyplot as plt


#image = Image.open("/content/drive/MyDrive/SRCNN_pytorch2/test/Lr_Copia de Img000000.tif")
#background = Image.open("/content/drive/MyDrive/Super Resolution/Average_Qc_0p10mLminQ_d0p05Lmin_freq1000Hz.tif")

#pixels = np.array(image)

#pixelsBackground = np.array(background)

##output = pixels - pixelsBackground

#plt.imshow(output, cmap='gray', vmin=0, vmax=255)

def subtract_image_from_directory(image_path, directory, output_directory, hr = False, scale = 0):
    # Create the output directory if it doesn't exist
    os.makedirs(output_directory, exist_ok=True)

    # Open the image to subtract
    subtract_image = Image.open(image_path)
    subtract_pixels = np.array(subtract_image)

    # Get a list of image files within the directory
    image_files = [f for f in os.listdir(directory) if os.path.isfile(os.path.join(directory, f))]

    # Iterate through each image file
    for image_file in image_files:

        # Open the image
        image_path = os.path.join(directory, image_file)
        image = Image.open(image_path)

        pixels = np.array(image)

        # Subtract the image
        subtracted_pixels = pixels - subtract_pixels
        #print(pixels)
        #print(subtract_pixels)

        #print(subtracted_pixels)

        #plt.imshow(subtracted_pixels, cmap='gray', vmin=0, vmax=255)


        # Create a new PIL image from the subtracted pixels
        subtracted_image = Image.fromarray(subtracted_pixels)

        # Save the subtracted image in the output directory
        output_path = os.path.join(output_directory, image_file)
        subtracted_image.save(output_path)



# Example usage
subtract_image_path = "/home/daniel/Documents/Super Resolution/Preprocessed/Cropped_Qc_0p10mLminQ_d0p05Lmin_freq1000Hz/Average_Downsampledx2_Qc_0p10mLminQ_d0p05Lmin_freq1000Hz.tif"
input_directory = "/home/daniel/Documents/SRCNN_pytorch2/averageTest"
output_directory = "/home/daniel/Documents/SRCNN_pytorch2/averageTest/result"
subtract_image_from_directory(subtract_image_path, input_directory, output_directory)

```
